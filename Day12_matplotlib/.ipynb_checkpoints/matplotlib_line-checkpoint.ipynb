## matplotlib and usage - for data visualisation

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
x=[1,2,3,4,5]
```

```python
y=[10,20,15,40,50]
```

#### Basic Matplotlib Chart
       - Pyplot : Interface for creating and managing plot
       - Figure : The complete window to create plot
       - Axes   : The area where data is plotted
       - Axis   : x-line and y-line to define scale and ticks
       - Plot   : The graph of data

```python
plt.figure(figsize=(8,6))
```

### marker
- `'o'`: Circle
- `'.'`: Point
- `','`: Pixel
- `'v'`: Triangle down
- `'^'`: Triangle up
- `'<'`: Triangle left
- `'>'`: Triangle right
- `'s'`: Square
- `'p'`: Pentagon
- `'*'`: Star
- `'+'`: Plus
- `'x'`: X
- `'D'`: Diamond
- `'d'`: Thin diamond
- `'|'`: Vertical line
- `'_'`: Horizontal line

For the `linestyle` parameter, common values include:
- `'-'`: Solid line
- `'--'`: Dashed line
- `'-.'`: Dash-dot line
- `':'`: Dotted line
- `'None'`
- `' '` 
- `''`
- `'solid'`
- `'dashed'`
- `'dashdot'`
- `'dotted'`
   

```python
plt.plot(x,y,marker='v',linestyle='-')
```

```python
plt.xlabel('X label')
```

```python
plt.ylabel('Y label')
```

```python
plt.title("Title")
```

```python
plt.grid(True)
```

#### Example 1
     - Create a line chart from students.xlxs
     - Generate line chart based on student marks and students name

```python
df = pd.read_excel("students.xlsx")
```

```python
name = df['Name']
marks = df['Marks']
```

```python
plt.plot(name, marks, marker='D',linestyle='--', color='brown' ,linewidth=2)
plt.title("Student Marks Line Chart")
plt.xlabel("Student Name")
plt.ylabel("Student Marks")
plt.grid(True)
```

```python
plt.figure(figsize=(10,6))
plt.plot(name, marks, marker='P',linestyle=':', color='purple' ,linewidth=2)
plt.title("Student Progress Line Chart")
plt.xlabel("Student Name")
plt.ylabel("Student Marks")
plt.grid(True)
plt.show()
```
