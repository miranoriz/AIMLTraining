Scatter Plot - represents data in points, where each points represent data observation

```python
import matplotlib.pyplot as plt
```

```python
product_sales = {'Laptop':750,'Smart Phone':650,'Smart Watch':400,'Tablet':500, 'AC':400}
products =list(product_sales.keys())
sales = list(product_sales.values())

plt.scatter(products,sales)
plt.title("Basic Porduct and Sales Representation")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
players = ['MSD','Rohit','Zeera','Jay','Izat']
match1_score = [45,67,56,78,89]
match2_score = [55,78,40,80,97]
match3_score = [60,95,12,25,100]
plt.figure(figsize=(10,5))
plt.scatter(players,match1_score,color='red',label='Match One',s=100) #s mean size of marker
plt.scatter(players,match2_score,color='blue',label='Match Two',s=100, marker='d')
plt.scatter(players,match3_score,color='green',label='Match Three',s=100, marker='p')
plt.title("Comparison of Players in Three Matches")
plt.xlabel('Player')
plt.ylabel('Scores')
plt.legend(loc = 'lower left') #show label
plt.grid(True,linestyle=':',alpha=0.5) #alpha ranges from 0.0 to 1.0: : transparent to opaque
plt.show()
```

```python
import pandas as pd
df = pd.read_csv("final_data.csv")
df.head()
```

```python
print(df['Workout_Type'])
```

```python
print(df['Calories_Burned'])
```

```python
avg_calories = df.groupby('Workout_Type')['Calories_Burned'].mean()
print(avg_calories)
```

```python
#show average calories burned for each workout type)
plt.figure(figsize=(10,6))
plt.scatter(avg_calories.index,avg_calories.values)
plt.title("Average Calories Burned by Workout Type")
plt.xlabel("Type of Workout")
plt.ylabel("Calories Burned")
plt.grid(True, linestyle = "--")
plt.show()
```
