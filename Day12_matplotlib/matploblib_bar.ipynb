```python
products = ['laptop','ipad','iphone','samsung','pc']
weekly_Sales = [500,200,1000,700,1000]
plt.bar(products,weekly_Sales)
plt.xlabel("Products")
plt.ylabel("Weekly Sales")
plt.show()
```

```python
products = ['laptop','ipad','iphone','samsung','pc']
weekly_Sales = [500,200,1000,700,1000]
plt.figure(figsize=(10,5))
bar_colors = ['purple','orange','brown','yellow','pink']
plt.bar(products,weekly_Sales,color=bar_colors,width=0.5,edgecolor='black')
plt.title("Weekly Product Sales")
plt.xlabel("Products")
plt.ylabel("Weekly Sales")
plt.show()
```

#### Exercise 1
     Represent data using bar chart
     x-label = product name
     y-label = price
     Use color for bars
     Use appropriate headings

```python
product_sales = {'Laptop':750,'Smart Phone':650,'Smart Watch':400,'Tablet':500, 'AC':400}
products =list(product_sales.keys())
sales = list(product_sales.values())

plt.figure(figsize=(10,5))
bar_colors = ['purple','orange','brown','yellow','pink']
plt.bar(products,sales,color=bar_colors,width=0.5,edgecolor='black')
plt.title("Product Sales")
plt.xlabel("Product Name")
plt.ylabel("Sales Value")
plt.show()
```

```python
import pandas as pd
df = pd.read_csv("patients.csv")
df.head(100)
```

```python
age_counts = df['age'].value_counts().sort_index()
plt.figure(figsize=(13,5))
plt.bar(age_counts.index,age_counts.values,color='skyblue',width=0.5,edgecolor='black')
plt.title("Number of Patient by Age")
plt.xlabel("Age")
plt.ylabel("Number of Patients")
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()
```

```python
print(df['service'].value_counts())
```

```python
#Finds the average satisfaction score for each service type.
avg_satisfaction= df.groupby('service')['satisfaction'].mean().sort_values(ascending=False)
avg_satisfaction
```

##### Create bar plot - Average Age of patients by arrival date

```python
#Finds the average age by arrival date.
avg_age_by_date= df.groupby('arrival_date')['age'].mean()
print(avg_age_by_date)
```

```python
plt.figure(figsize=(13,5))
plt.bar(avg_age_by_date.index, avg_age_by_date.values, color='pink', width=0.5, edgecolor='pink')
plt.title("Average age of Patients by Arrival Date")
plt.xlabel("Arrival Date")
plt.ylabel("Average Age")
plt.show()
```
