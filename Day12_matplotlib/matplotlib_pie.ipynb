```python
import matplotlib.pyplot as plt
```

```python
products = ['laptop','ipad','headphone','keyboard']
weekly_sales = [400,300,200,100]
```

```python
plt.figure(figsize=(10,5))
plt.pie(weekly_sales,labels=products)
plt.title("Weekly Product Sales")
plt.show()
```

```python
product_sales = {'Laptop':750,'Smart Phone':650,'Smart Watch':400,'Tablet':500, 'AC':400}
products =list(product_sales.keys())
sales = list(product_sales.values())
plt.figure(figsize=(10,5))
pie_colors = ['deepskyblue','dodgerblue','lightskyblue','cornflowerblue','powderblue']
plt.pie(sales,labels=products,colors=pie_colors,shadow=True, startangle=45)
plt.title("Product Sales")
plt.show()
```

```python
import pandas as pd
df = pd.read_csv("patients.csv")
df
```

```python
services_counts = df["service"].value_counts()
print(services_counts)
```

```python
service_name1 = pd.unique(df['service'])
print(service_name1)
```

```python
service_name = services_counts.index
print(service_name)
```

```python
plt.pie(services_counts, labels=service_name1, shadow=True)
plt.title("Distribution of Patients by service", fontsize=20, fontweight='bold')
plt.show()
```
