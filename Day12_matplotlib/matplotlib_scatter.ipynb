Scatter Plot - represents data in points, where each points represent data observation

```python
import matplotlib.pyplot as plt
```

```python
product_sales = {'Laptop':750,'Smart Phone':650,'Smart Watch':400,'Tablet':500, 'AC':400}
products =list(product_sales.keys())
sales = list(product_sales.values())

plt.scatter(products,sales)
plt.title("Basic Porduct and Sales Representation")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
