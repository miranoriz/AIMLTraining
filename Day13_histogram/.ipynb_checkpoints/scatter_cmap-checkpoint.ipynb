```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
temp = np.random.randint(20,30,size=50)
#inferno, cool, rainbow, gray, viridis, hot
plt.scatter(x,y,c=temp, cmap='plasma', s=120,edgecolor='black')
plt.colorbar(label='Temperature')
plt.title("Scatter Plot Coloredby Temperature")   
plt.xlabel("x-random")
plt.ylabel("y-label")
plt.show()
```

```python
height_cm =[150,155,160,165,170,175,180,185]
weight_kg = [50,53,56,59,62,66,68,72]
age_year = [20,22,25,28,30,35,38,40]
```

```python
plt.figure(figsize=(10,6))
scatter=plt.scatter(height_cm , weight_kg, c=age_year , cmap='rainbow', s=120, alpha=0.9)
plt.colorbar(label= 'Age (years)')
plt.xlabel('Weight (KG)')
plt.ylabel('Height (CM)')
plt.grid('b--', linewidth =0.7)
plt.show()
```
