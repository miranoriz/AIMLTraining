### Histogram - shows how numerical data is distributed

```python
import matplotlib.pyplot as plt
```

```python
marks = [15,10,15,10,25,20,25,40,45,40,55,60,65,60,75,90,95,90,95,90]
plt.style.use("ggplot")
plt.figure(figsize=(10,5))
plt.hist(marks, bins=5, color='khaki', edgecolor='darkblue', alpha=0.8)
plt.title("Histogram of Student Marks")
plt.xlim(5,100) 
plt.xlabel("Marks Range")
plt.ylabel("Number of Students")
plt.show()
```
