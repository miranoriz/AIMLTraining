```python
import matplotlib.pyplot as plt
```

```python
depts = ['HR', 'Web-dev','App-Dev','Agile-Scrum','Testing']
expenses = [12000,25000,18000,20000,25000]
```

```python
plt.figure(figsize=(10,5))
plt.pie(expenses, labels=depts, autopct='%1.1f%%', startangle=140, explode =(0,0.05,0,0,0),shadow=True)
plt.title('Department wise Expense Report')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.pie(expenses, labels=depts, colors= plt.cm.inferno([0.9,0.8,0.5,0.7,0.6]),autopct='%1.1f%%', startangle=140,shadow=True, radius= 1.2)
plt.title('Department wise Expense Report \n\n')
plt.show()
```

```python
sales_data = {'mobile':45000, 'clothing':30000, 'groceries':35000,'home decor':20000}
```

```python
goods = list(sales_data.keys())
sales = list(sales_data.values())
```

```python
plt.figure(figsize=(10,6))
pie_colors = ['lightblue','lightpink','lightgray','lightyellow']
plt.pie(sales,labels=goods, autopct='%1.2f%%', textprops={'color':'black','fontsize':10,'fontweight':'bold'},colors= pie_colors, startangle=140, explode =(0,0.1,0,0),shadow=True)
plt.title('Goods Sales Report')
plt.savefig("sa.png", dpi=300, bbox_inches="tight")
plt.show()
```
