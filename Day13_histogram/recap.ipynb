```python
import matplotlib.pyplot as plt
```

```python
month = ['jan','feb','mar','apr']
sales = [2000,3000,2500,3600]
```

```python
plt.figure(figsize=(10,5))
plt.plot(month,sales,'c:', marker='d',linewidth=2, label='Monthly Sales')
plt.title("Monthly Sales Report")
plt.xlabel('Months')
plt.ylabel('Sales')
plt.xlim(0,3)
plt.ylim(2000,4000)
plt.grid(True)
plt.legend(loc='upper left',fontsize=10)
plt.show()
```

```python
import numpy as np
```

```python
x=np.linspace(0,2,100) 
#The linspace function is great when you want to draw smooth curves
#containing 100 evenly spaced numbers between 0 and 2
plt.plot(x,x,label='Linear')
plt.plot(x,x**2,label='Quadratic')
plt.plot(x,x**3,label='Cubic')
plt.xlabel("X lines")
plt.ylabel("Y lines")
plt.title("Simple Lines Plot")
plt.legend(loc='upper left')
plt.show()
```

```python
categories = ["Electronics","Mobiles", "Books","Groceries","Home Decoration"]
sales1 = [5000,6000,3000,9000,4000]
plt.figure(figsize=(7,5))
plt.bar(categories,sales1,color='pink',linewidth=1,edgecolor='black',label='Sales Info')
plt.title("Goods Sales Report")
plt.xlabel('Categories')
plt.ylabel('Sales')
plt.grid(axis='y',linestyle=':',alpha=0.7,color='blue')
plt.tight_layout()
plt.xticks(rotation=20,fontsize=10,color='black')
plt.yticks([2000,4000,6000,8000,10000],['20k','40k','60k','80k','1M'],fontsize=12, color='green')
plt.legend(title="Info",fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(9,3))
bar_colors = ['skyblue','navy']
plt.barh(month,sales,color=bar_colors,linewidth=2, edgecolor = 'black')
plt.title("Monthly Sales Report")
plt.xlabel('Sales')
plt.ylabel('Months')
plt.show()
```

```python
height_cm =[150,155,160,165,170,175,180,185]
weight_kg = [50,53,56,59,62,66,68,72]
plt.figure(figsize=(9,3))
plt.scatter(height_cm,weight_kg,s=200, c='pink',marker='*', alpha=0.9, label='weight/height', edgecolor = 'black')
plt.title("Scatter Plot Weight VS Height")
plt.xlabel('Height(CM)')
plt.ylabel('Weight(KG)')
plt.xlim(145,187)
plt.ylim(40,90)
plt.xticks(rotation=20,fontsize=10,color='green')
plt.yticks([50,55,60,65,70,75,80,85],['50KG','55KG','60KG','65KG','70KG','75KG','80KG','85KG'],fontsize=8, color='green')
plt.grid(color='lightblue',linestyle=":")
plt.legend(loc='upper left', fontsize=9)
plt.show()
```
