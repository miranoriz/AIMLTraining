```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("/Users/aziim/Desktop/AIMLTraining/Day13_histogram/netflix_data.csv")
df.head()
```

```python
sns.set_theme(style='white')
sns.countplot(x='type', data=df, hue='type')
plt.title("Number of Movie vs Number of TV Shows")
plt.xlabel("Content Type")
sns.despine()
plt.show()
```

```python
df=df.dropna(subset=['type','genre'])
df['type']
```

```python
df['genre']
```

```python
df['country']
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x='type', data=df.head(30), hue='country',palette='Set2')
plt.title("Netflix Content Type by Country")
plt.xlabel("Content Type")
plt.legend(title='Country',bbox_to_anchor=(1.02,1),loc='upper left')
sns.despine()
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x='type', data=df, hue='country',palette='Set3')
plt.title("Netflix Content Type by Country")
plt.xlabel("Content Type")
plt.legend(title='Country',bbox_to_anchor=(1.02,1),loc='upper left',ncol=4, frameon=True)
sns.despine()
plt.savefig("netflix_country.jpg",dpi=300, bbox_inches=("tight"))
plt.show()
```

Using netflix dataset to visualize the distribution of release year for movie and tv show using histplot()

Wrong Example

sns.histplot() doesn’t accept a list of arrays like plt.hist() does.
sns.histplot() bar can be different color using hue
figure 1 can only 1 color since no hue

```python
plt.figure(figsize=(10,5))
movie = df[df['type']=='Movie']
shows = df[df['type']=='TV Show']
sns.histplot([movie['release_year'],shows['release_year']],bins=50,palette='Set2', edgecolor='black')
plt.title("Netflix Release Title by Year Figure 1")
plt.xlabel("Release Date")
plt.ylabel("Number of Title")
plt.grid('b--',linewidth=0.2)
plt.show()
```

```python
#Correct Example

plt.figure(figsize=(10,5))
movie = df[df['type']=='Movie']
shows = df[df['type']=='TV Show']
sns.histplot(data=df,x="release_year",hue="type",palette=["lightpink", "skyblue"], multiple="dodge",bins=30,edgecolor="white")
plt.title("Netflix Release Title by Year")
plt.xlabel("Release Date")
plt.ylabel("Number of Title")
plt.grid('b--',linewidth=0.2)
plt.show()
```
