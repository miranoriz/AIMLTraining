## subplot()
- provide a way to plot multiple line on single figure
- given numbers of row and column. it return tuple (fig,ax) giving a single figure with an array of axes

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
sales_data = {'Electronics':45000,'clothing':30000,'grocery':25000, 'mobile':20000}
category = list(sales_data.keys())
sales = list(sales_data.values())
```

```python
fig,((ax1,ax2),(ax3,ax4),(ax5,ax6))=plt.subplots(nrows=3,ncols=2,figsize=(10,10))
x=np.random.random(10)
ax1.plot(x,x/2,'g--')
ax2.scatter(np.random.random(10),np.random.random(10), c='red')
ax3.bar(category,sales,color=['pink','cyan'])
ax4.pie(sales, labels=category, autopct= '%1.1f%%')
ax5.hist(np.random.randn(100),color='khaki',edgecolor='black')
ax6.scatter(category,sales, color='magenta', marker ='p',s=100)
```

#### subplots using indexing

```python
our_lines = np.linspace(0,10,100)
```

```python
product_prices = {'Hair Oil':20, 'Face Wash':30, 'Lip Balm':10, 'Face Cream':15}
print(list(product_prices.keys()))
print(product_prices.keys())
```

```python
our_number = np.random.random(10)
```

```python
our_dots = np.random.random(100)
```

```python
fig,ax=plt.subplots(nrows=3,ncols=2,figsize=(10,10))
ax[0,0].plot(our_lines,our_lines/2, 'r:')
ax[0,1].scatter(np.random.random(100),np.random.random(100),c='navy')
ax[1,0].bar(list(product_prices.keys()),list(product_prices.values()),color=['pink','cyan','lightgrey','yellow'],edgecolor='black')
ax[1,1].pie(list(product_prices.values()), labels=list(product_prices.keys()), autopct= '%1.1f%%')
ax[2,0].hist(np.random.randn(100),color='khaki',edgecolor='black')
ax[2,1].scatter(list(product_prices.keys()),list(product_prices.values()), color='magenta', marker ='p',s=100)
fig.suptitle("Subplot Examples ")

ax[0,0].set_title("Line Plot Example")
ax[0,0].grid(alpha=0.2)
ax[0,0].set_xlabel("x-label")
ax[0,0].set_ylabel("y-label")

ax[0,1].set_title("Scatter Random Example")
ax[0,1].grid(alpha=0.2)
ax[0,1].set_xlabel("x-label")
ax[0,1].set_ylabel("y-label")

ax[1,0].set_title("Bar Plot Example")
ax[1,0].grid(alpha=0.2)
ax[1,0].set_xlabel("x-label")
ax[1,0].set_ylabel("y-label")

ax[1,1].set_title("Pie Plot Example")

ax[2,0].set_title("Histogram Plot Example")
ax[2,0].grid(alpha=0.2)
ax[2,0].set_xlabel("x-label")
ax[2,0].set_ylabel("y-label")

ax[2,1].set_title("Scatter Plot Example")
ax[2,1].grid(alpha=0.2)
ax[2,1].set_xlabel("x-label")
ax[2,1].set_ylabel("y-label")

plt.tight_layout()
plt.savefig("subplotex.png",dpi=300,bbox_inches="tight")
```
