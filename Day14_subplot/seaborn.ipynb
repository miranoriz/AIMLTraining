```python
data = sns.load_dataset('iris')
print(data)
bar_colors = ['thistle','mediumvioletred','orchid']
sns.barplot(x='species',y='sepal_width', data=data, hue='species',palette=bar_colors)
plt.title('Iris Species Distribution Bar Chart')
plt.xlabel('Name Species')
plt.ylabel('Sepal Width')
plt.grid(axis='y',linestyle='--',color='lightblue', alpha=0.5)
plt.show()      
```

```python
sns.violinplot(x='species',y='sepal_width',data=data,hue='species',palette=bar_colors)
plt.title('Iris Species Distribution Bar Chart')
plt.xlabel('Name Species')
plt.ylabel('Sepal Width')
plt.grid(axis='y',linestyle='--',color='lightblue', alpha=0.5)
```

```python
sns.violinplot(x='species',y='sepal_width',data=data,hue='species',palette='Set3')
plt.title('Iris Species Distribution Bar Chart')
plt.xlabel('Name Species')
plt.ylabel('Sepal Width')
plt.grid(axis='y',linestyle='--',color='lightblue', alpha=0.5)
```

```python
sns.stripplot(x='species',y='sepal_width',data=data,hue='species',palette='Set1')
plt.title('Iris Species Distribution Bar Chart')
plt.xlabel('Name Species')
plt.ylabel('Sepal Width')
plt.grid(axis='y',linestyle='--',color='lightblue', alpha=0.5)
```

```python
sns.pairplot(data=data,hue='species')
plt.title('Iris Species Distribution Bar Chart')
plt.xlabel('Name Species')
plt.ylabel('Sepal Width')
plt.grid(axis='y',linestyle='--',color='lightblue', alpha=0.5)
```

```python
sns.set_style('dark')
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=data)
plt.title("Darkgrid Example")
plt.show()
```

```python
sns.set_style('ticks')
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=data)
plt.title("Ticks Example")
plt.xlabel("Sepal Length in CM")
plt.ylabel("Sapel Width in CM")
plt.xticks([4,5,6,7,8],fontsize=10,color="darkblue")
plt.yticks([1,2,3,4,5,6,7],fontsize=10,color='fuchsia')
sns.despine() #remove border line
plt.show()
```

```python
sns.set_style('whitegrid')
sns.boxplot(x='sepal_length',y='sepal_width',hue='species',data=data)
plt.title("Ticks Example")
plt.show()
```

```python
sns.set_style('whitegrid')
sns.histplot(x='sepal_length',y='sepal_width',hue='species',data=data)
plt.title("Ticks Example")
plt.show()
```

```python
sns.set_style('whitegrid')
sns.kdeplot(x='sepal_length',y='sepal_width',hue='species',data=data)
plt.title("KDE Example")
plt.show()
```

```python
sns.set_style('white')
sns.lineplot(x='sepal_length',y='sepal_width',data=data)
plt.title("Line Plot Example")
plt.show()
```

```python
sns.set_style('white')
sns.swarmplot(x='species',y='sepal_width',hue='species',data=data)
plt.title("Ticks Example")
plt.show()
```

```python
sns.countplot(x='species', data=data, palette='pastel', hue='species')
plt.title('Number of Samples per Iris Species')
plt.show()
```

```python
sns.countplot(y='species', data=data, palette='pastel', hue='species')
plt.title('Number of Samples per Iris Species')
plt.show()
```
