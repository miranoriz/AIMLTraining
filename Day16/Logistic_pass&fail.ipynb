Scenario where there is two output (Fail/Pass)

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("pass_fail.csv")
df
```

```python
x = df[['Hours']]
y = df['Result']
```

```python
model = LogisticRegression()
model.fit(x,y)
```

```python
new_hour = input('Enter New Hour of Study')
predicted_result = model.predict(pd.DataFrame([[new_hour]], columns=['Hours']))
print(f"The predicted result for {new_hour} hours of Study is {predicted_result[0]}")                                 
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x, y, c='black', label='Actual Data', s=80)
plt.plot(x, model.predict(x), color='orange', linewidth=2, label='Regression Line')
plt.scatter(new_hour, predicted_result, color='magenta', s=100, edgecolor='black', label='Predicted Point')
plt.title("Hours of Study vs Result", fontsize=14)
plt.xlabel("Study Hours")
plt.ylabel("Result")
plt.legend()
plt.show()
```
