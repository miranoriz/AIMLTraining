```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'experience_year':[2,4,5,3,6,5],
    'performance_rating':[3,4,5,3,4,4],
    'promotion':['no','yes','yes','no','yes',None]
}

data = pd.DataFrame(data)
data
```

```python
#clean data
df = data.dropna()
df
```

```python
x = df[['experience_year','performance_rating']]
y = df['promotion']
```

```python
model = DecisionTreeClassifier()
model.fit(x,y)
```

```python
new_exp = int(input("Enter Staff Experience:"))
new_perf = int(input("Enter Performance Rating:"))
pred_pro = model.predict(pd.DataFrame([[new_exp, new_perf]], columns=['experience_year', 'performance_rating']))
print(f'\nExperience year: {new_exp} \nPerformance rating: {new_perf} \nPromotion prediction: {pred_pro[0]}')
```

```python
plt.figure(figsize=(8,5))
plot_tree(model, feature_names=['experience_year', 'performance_rating'], class_names=['no','yes'],filled=True)
plt.title("Decision Tree - Employee Promotion")
plt.tight_layout()
plt.show()
```

```python
data = {
    'Age': [25,30,45,35,22,40,50],
    'Income': ['Low','Medium','High','Medium','Low','High','High'],
    'Buy': ['No','No','Yes','Yes','No','Yes','Yes']
}
data1 = pd.DataFrame(data)
data1
```

```python
ohe = OneHotEncoder(sparse_output=False)
income_encoded= pd.DataFrame(ohe.fit_transform(data1[['Income']]), columns=ohe.get_feature_names_out(['Income']))
print(income_encoded)
```

```python
data = data1.drop('Income',axis=1)
income_encoded_d = pd.DataFrame(income_encoded, columns=ohe.get_feature_names_out(['Income']))  # Create DataFrame with same index as data

x = pd.concat([data[['Age']], income_encoded_d], axis=1)
y = data['Buy'] 
```

```python
model = DecisionTreeClassifier()
model.fit(x,y)
```

```python
new_age = int(input("Enter Age:"))
new_income = input("Enter Income Level:")
new_income_df = pd.DataFrame([new_income], columns=['Income']) 
income_encode = ohe.transform(new_income_df)

income_encoded_df = pd.DataFrame(income_encode, columns=ohe.get_feature_names_out(['Income']))
new_data = pd.concat([pd.DataFrame({'Age': [new_age]}), income_encoded_df], axis=1)

pred_buy = model.predict(new_data)
print(f'\nAge Year: {new_age} \nIncome Level: {new_income} \nBuy or Not: {pred_buy[0]}')
```

```python
plt.figure(figsize=(8,5))
plot_tree(model, feature_names=['Age', 'Income'], class_names=['no','yes'],filled=True)
plt.title("Decision Tree - Buy Decision")
plt.tight_layout()
plt.show()
```
