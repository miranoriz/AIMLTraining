```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("house_price.csv")
df = df.head(5)
df
```

```python
x = df[['Size_sqft']]
y = df['Price']
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
new_sqft = float(input('Enter New Size of House'))
predicted_price = model.predict(pd.DataFrame([[new_sqft]], columns=['Size_sqft']))
# predicted_price1 = model.predict(pd.DataFrame([[1700]], columns=['Size_sqft']))
print(f"The predicted house price for size {new_sqft} sqft is RM {predicted_price[0]:.2f}")
# print(f'{predicted_price1[0]:.2f}')
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x, y, c='fuchsia', label='Actual Data', s=80)
plt.plot(x, model.predict(x), 'y-', label='Regression Line')
plt.scatter(new_sqft, predicted_price, c='olive', s=80, label='Predicted Point')
plt.title("House Size vs Price")
plt.xlabel("House Size (sqft)") 
plt.ylabel("Price (RM)")
plt.grid(alpha=0.2)
plt.legend()
plt.show()
```
