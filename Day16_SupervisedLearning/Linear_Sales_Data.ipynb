### Supervised Machine Learning

#### 1. Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("sales_Data.csv")
df
```

```python
x = df[['Advertising_Spend']]
y = df['Sales_Revenue']
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
newspend = float(input('Enter New Spend Revenue RM'))
predicted_revenue = model.predict(pd.DataFrame([[newspend]], columns=['Advertising_Spend']))
#predicted_revenue[0] -> extract a value from the array
print(f"New Predcited for RM {newspend} is RM {predicted_revenue[0]:.2f}")
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x, y, c='deeppink', label='Actual Data', s=80)
plt.plot(x, model.predict(x), 'g:', label='Regression Line')
plt.scatter(newspend, predicted_revenue, c='navy', label='Predicted Data', s=80)
plt.title("Advertising Spend vs Sales Revenue")
plt.xlabel("Advertising Spend (RM)")
plt.ylabel("Sales Revenue (RM)")
plt.legend()
plt.grid(alpha=0.2)
plt.show()
```
