```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from math import sqrt
import matplotlib.pyplot as plt
```

```python
data_set={ 'Adv_Spend':[5,10,15,20,25,30,35,40,45,50,55,60], 'Revenue_Gain':[350,370,400,430,450,480,500,520,550,580,600,630] }
df = pd.DataFrame(data_set)
df
```

```python
x = df[['Adv_Spend']]
y = df['Revenue_Gain']
model = LinearRegression()
model.fit(x,y)
```

```python
pred_rev = model.predict(x)

mae = mean_absolute_error(y, pred_rev) #General accuracy check
mse = mean_squared_error(y,pred_rev) #Model training, optimization
rmse = sqrt(mse) #Evaluation, comparison
r2=r2_score(y,pred_rev) # measures how well your model explains the variation in the actual data.

print('Mean Absolute Error!',round(mae,2))
print('Mean Squared Error!',round(mse,2))
print("Root Mean Squared Error!",round(rmse,2))
print("R-Squared Score is:",round(r2,2))
```

```python
newspend = int(input('Enter New Ads Spend RM'))
predicted_revenue = model.predict(pd.DataFrame([[newspend]], columns=['Adv_Spend']))
print(f"Predicted revenue for ads spend for RM {newspend} is RM {predicted_revenue[0]:.2f}")
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x,y, color='deeppink', label='Actual Revenue', s=100)
plt.plot(x, pred_rev, 'g:', label='Prediction Line')
plt.scatter(newspend, predicted_revenue, c='navy', label='Predicted Revenue', s=80)

plt.title('Actual Revenue vs Predicted Revenue')
plt.xlabel('Ads Spend')
plt.ylabel('Revenue')
plt.legend()
plt.grid(True)
plt.show()
```
