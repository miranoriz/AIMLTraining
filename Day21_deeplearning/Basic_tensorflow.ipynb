```python
import tensorflow as tf
```

```python
print (tf.__version__)
```

```python
#Single number is scalar
scalar = tf.constant(7)
scalar
```

```python
scalar.ndim
```

```python
# number with direction
vector = tf.constant([10,10,20,30])
vector
```

```python
vector.ndim
```

```python
matrix = tf.constant([[10,10],[10,11]])
matrix.ndim
```

```python
#2D array numbers
second_matrix = tf.constant([[10,10],[10,11],[10,12],[10,13],[10,14]])
second_matrix.ndim
```

```python
third_matrix = tf.constant([[[10,10],[10,11]],[[10,13],[10,13]]])
print(third_matrix)
third_matrix.ndim
```

```python
second_matrix=tf.constant([[5.6,7.3], [3.45,8.25],[3.11,9.23],[4.34,2.34],[8.4,6.23]], dtype=tf.float32)
print(second_matrix)
```

```python
# Tensor : n-dimensional array of number

tensor=tf.constant ([[[1,2], [3,4]], [[5,6], [7,8]]])
tensor
```

```python
#Find the shape,rank and size of tensor
print("Scalar Information")
print("Shape:",scalar.shape)
print ("Dimension:",scalar.ndim)
print("Rank:",tf.rank(scalar).numpy())
print("Size:",tf.size(scalar))
```

```python
print("Vector Information")
print("Shape:",vector.shape)
print ("Dimension:",vector.ndim)
print("Rank:",tf.rank(vector).numpy())
print("Size:",tf.size(vector))
```

```python
print("Matrix Information")
print("Shape:",matrix.shape)
print ("Dimension:",matrix.ndim)
print("Rank:",tf.rank(matrix).numpy())
print("Size:",tf.size(matrix))
```

```python
print("Tensor Information")
print("Shape:",tensor.shape)
print ("Dimension:",tensor.ndim)
print("Rank:",tf.rank(tensor).numpy())
print("Size:",tf.size(tensor))
```
