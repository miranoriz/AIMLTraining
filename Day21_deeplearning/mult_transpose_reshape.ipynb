```python
import tensorflow as tf
import numpy as np
```

**multiply by column [m x n]*[n x P] = [m x p]**

```python
tensor_a = tf.constant([[1,2],[3,4],[6,7]],dtype=tf.int32)
tensor_b = tf.constant([[5,6],[7,8]],dtype=tf.int32)

print("Shape of A:",tensor_a.shape)
print("Shape of B:",tensor_b.shape)
```

```python
tensor_c = tf.matmul(tensor_a, tensor_b)
print(tensor_c)
print("Shape of C:",tensor_c.shape)
```

```python
x = tf.constant([[1,0,-1],[2,3,1]],dtype=tf.float32)
y = tf.constant([[4,],[5,],[6,]],dtype=tf.float32)
print("Shape of x:",x.shape)
print("Shape of y:",y.shape)
```

```python
z = tf.matmul(x, y)
print(z)
print("Shape of Z:",z.shape)
```

**Transpose**

```python
a = tf.constant([[[1,0,-1],[2,3,1]],[[3,4,5],[6,7,8]]],dtype=tf.int32)
print(a)
print("Shape of a:",a.shape)
```

```python
b = tf.transpose(a)
print(b)
print("Shape of b:",b.shape)
```

```python
#swap row to column in a block
c = tf.transpose(a,perm=[0,2,1])
print(c)
print("Shape of c:",c.shape)
```
