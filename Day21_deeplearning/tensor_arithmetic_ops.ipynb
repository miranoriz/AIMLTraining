```python
import tensorflow as tf
import numpy as np
```

```python
tensor_one = tf.constant([[11,9],[2,5]])
tensor_one
```

```python
#add
tensor_one + 15
tf.add(tensor_one,15)
```

```python
#sub
tensor_one - 3
tf.subtract(tensor_one,3)
```

```python
#mult
tensor_one * 3
tf.multiply(tensor_one,3)
```

```python
#div
tensor_one / 2
tf.divide(tensor_one,2)
```

```python
#square
tf.square(tensor_one)
```

```python
#square root, #cast -> change datatype
tf.sqrt(tf.cast(tensor_one, dtype=tf.float32))
```

Exercise Create two tensors with random values between 0 and 1, shape is  
[6,300] **Find out the following details: **  
• Number of elements  
• Size,  
• Rank,  
• Shape  

```python
random_tensor_one = tf.random.uniform(shape=[6, 300], minval=0, maxval=1)
random_tensor_two = tf.random.Generator.from_seed(42).normal(shape=(6,300))

print("First Tensor Details")
print("Number of element:",tf.size(random_tensor_one).numpy())
print("Size:",random_tensor_one.shape)
print("Rank:",random_tensor_one.ndim)
print("Shape:",random_tensor_one.shape)

print("\nSecond Tensor Details")
print("Number of element:",tf.size(random_tensor_two).numpy())
print("Size:",random_tensor_two.shape)
print("Rank:",random_tensor_two.ndim)
print("Shape:",random_tensor_two.shape)
```
