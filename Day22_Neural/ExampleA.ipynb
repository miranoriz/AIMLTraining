Building a regression model that predicts medical charges, using a simple fully connected neural network trained with Stochastic Gradient Descent to minimize Mean Absolute Error.

```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('/content/insurance_data.csv')
data
```

```python
ins_data =pd.get_dummies(data)
ins_data
```

```python
x = ins_data.drop('charges', axis=1)
y = ins_data['charges']
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
tf.random.set_seed(42)
model = tf.keras.Sequential([tf.keras.layers.Dense(1),
                             tf.keras.layers.Dense(1)])
```

```python
model.compile(loss=tf.keras.losses.mae,
              optimizer='adam',
                metrics=['mae'])
```

```python
history =model.fit(x_train,y_train, epochs=200, verbose=0)
```

```python
model.evaluate(x_test,y_test) #the output will be total loss and mae
```

```python
plt.plot(history.history['loss'], label= 'loss(MSE)')
plt.legend()
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```
