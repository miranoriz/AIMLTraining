
(1) **Binary classification**

Predict whether or not someone has heart disease based on their health parameters. This is called binary classification since there
are only two options.

(2) **Multi-class classification**

Decide whether a photo of is of food, a person or a dog. This is called multi-class classification since there are more than two
options. e.g.Predict whether an image belongs to one of 10 categories (e.g., airplane, car, cat, dog, etc.)

(3) **Multi-label classification**

Predict what categories should be assigned to a Wikipedia article. This is called multi-label classification since a single article
could have more than one category assigned. OR Predict multiple genres (Action, Drama, Comedy, etc.) for a movie based on its
description in Netflix.
