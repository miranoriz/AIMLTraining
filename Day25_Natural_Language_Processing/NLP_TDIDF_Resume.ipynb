```python
#Import All Required Libraries
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import vectorize
```

```python
data = pd.read_csv("/content/UpdatedResumeDataSet.csv")
data
```

```python
print('\n Columns', data.columns.tolist())
```

```python
#random data
sample_df=data.sample(10, random_state=42).reset_index(drop=True)
sample_df
```

```python
sample_df=sample_df.rename(columns={'Resume': 'Resume_Text'})
sample_df

```

```python
vectorizer=TfidfVectorizer(stop_words='english', max_features=5000)
tfidf_matrix=vectorizer.fit_transform(sample_df ['Resume_Text'])
tfidf_matrix.shape
```

```python
similartiness_matrix=cosine_similarity(tfidf_matrix)
sim_df=pd.DataFrame(similartiness_matrix, index=sample_df['Category'])
sim_df
```

```python
resumes={
"candidate": ['Sam', 'Arsh', 'Maan','zoya'],
"resume_text": ["Python developer skilled in Django, Flask, Rest API and MySQL",
"Machine learning engineer experience in Python, Tensorflow, NLP and data anlysis",
"Full stack developer with .net,javascript,react,node.js,mongodb experience",
 "Full stack developer with .net,javascript,react,node.js,python"             ]
          }
```

```python
df = pd.DataFrame(resumes)
df
```

```python
vectorizer=TfidfVectorizer(stop_words='english', max_features=5000)
tfidf_matrix=vectorizer.fit_transform(df['resume_text'])
tfidf_matrix.shape
tfidf_matrix
```

```python
similartiness_matrix=cosine_similarity(tfidf_matrix)
sim_df=pd.DataFrame(similartiness_matrix, index=df["candidate"], columns=df ["candidate"])
sim_df.round(2)
```
