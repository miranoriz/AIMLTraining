```python
import pandas as pd, numpy as np, re
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score,classification_report, confusion_matrix
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
df = pd.read_csv('UpdatedResumeDataSet.csv')
df.head()
```

```python
import string
def clean_text(text):
  text=str(text)
  text=re.sub(r'http\S+','',text)
  text=text.translate (str.maketrans('','', string.punctuation))
  text=re.sub(r'\s+',' ', text)
  text=re.sub(r'\s+[a-zA-Z]\s+',' ',text)
  text = re.sub(r'[^\x00-\x7f]','', text)
  text=text.lower()
  return text
```

```python
df = pd.read_csv('UpdatedResumeDataSet.csv')
df
```

```python
df['clean_resume'] = df['Resume'].apply(clean_text)
display(df[['Resume', 'clean_resume']])
```

```python
x = df["clean_resume"]
y = df['Category']

tfidf = TfidfVectorizer(stop_words="english",max_features=3000)
matrix= tfidf.fit_transform(x)
matrix_df = pd.DataFrame(matrix.toarray(), columns=tfidf.get_feature_names_out())
matrix_df
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
vectorizer = CountVectorizer()
x_train = vectorizer.fit_transform(x_train)
x_test= vectorizer.transform(x_test)
```

```python
model=MultinomialNB()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
conf_matrix=confusion_matrix(y_test,y_pred)
report = classification_report(y_test, y_pred)
print("Accuracy:\n",accuracy)
# print("\nConfusion Matrix:\n",conf_matrix)
print("\nClassification Report:\n",report)
```

```python
resume_msg=[
     "Python developer required with strong coding skills ",
     " Experience in REST API, Automation, Jenkins, Flask and Python",
     "Looking for machine learning engineer familiar with NLP,Supervised, Unsupervised and Deep Learning ",
     "computer skills software knowledge mspower point"
     ]

new_resume_msg=[clean_text(msg) for msg in resume_msg]
new_msg = vectorizer.transform(new_resume_msg)
predictions = model.predict(new_msg)
print("Predictions are:\n")
for msg, prediction in zip(resume_msg, predictions):
    print(f"Message: {msg}\nPrediction: {prediction}\n")
```
