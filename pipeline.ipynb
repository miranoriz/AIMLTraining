```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
df = pd.DataFrame({
    "Study_Hours" : [1,2,3,4,5,np.nan,7,8,9,10],
    "attendance" : [60,65,70,75,80,85,np.nan,92,95,100],
    "final_mark" : [40,45,50,55,60,65,70,75,80,85]
})

df
```

```python
x = df [['Study_Hours', 'attendance']]
y = df['final_mark']
```

```python
#create the pipeline

pipe = Pipeline([
    ("imputer",SimpleImputer(strategy="mean")), #fill missing value with mean data
    ("scaler",StandardScaler()), #normalization data
    ("model",LinearRegression())# use model linear regression
])
pipe
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size = 0.2, random_state=42)
print("Training test size:",len(x_train))
print("Testing size:",len(x_test))
```

```python
pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
print(f"Predicted values are {y_pred}")
```

```python
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("Mean Absolute Error (MAE):" ,round(mae,2))
print(f"RootnMean Squared Error (MSE): {(round(rmse,2))}")
print(f"R-squared (R2): {(round(r2,2))}")
```

```python
Study_Hours = input("study`s hours: ")
attendance = input("attendance:")
new_student = pd.DataFrame({
    "Study_Hours":[Study_Hours],
    "attendance":[attendance]})
predict_marks = pipe.predict(new_student)
print(round(predict_marks[0],2))
```
